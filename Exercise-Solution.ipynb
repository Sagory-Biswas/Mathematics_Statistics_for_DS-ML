# Normal Distribution and Z Score for Outlier Removal

```python
ndz= pd.read_csv('bhp.csv')
ndz
```

```python
ndz.price_per_sqft.describe()
```

```python
ndz.price_per_sqft.quantile(0.001)
```

```python
ndz.price_per_sqft.quantile(0.999)
```

```python
ndz[ndz.price_per_sqft < 1366.184]
```

```python
ndz[ndz.price_per_sqft > 50959.362]
```

```python
ndz[(ndz.price_per_sqft < 1366.184)|(ndz.price_per_sqft > 50959.362)]
```

```python
NDZ = ndz[(ndz.price_per_sqft > 1366.184)&(ndz.price_per_sqft < 50959.362)]
NDZ
```

```python
NDZ.price_per_sqft.describe()
```

# using 4 standard deviation to remove outliers:

```python
mn = NDZ.price_per_sqft.mean()
mn
```

```python
st = NDZ.price_per_sqft.std()
st
```

```python
max_limit = mn + 4*st
max_limit
```

```python
min_limit = mn - 4*st
min_limit
```

```python
NDZ[NDZ.price_per_sqft>max_limit]
```

```python
NDZ[NDZ.price_per_sqft<min_limit]
```

```python
fndz= NDZ[(NDZ.price_per_sqft<max_limit)&(NDZ.price_per_sqft>min_limit)]
fndz
```

```python
fndz.shape
```

```python
NDZ.shape
```

```python
NDZ.shape[0] - fndz.shape[0]
```

# using zscore of 4 to remove outliers:

```python
NDZ.head()
```

```python
NDZ['z_score'] = (NDZ.price_per_sqft - NDZ.price_per_sqft.mean()) / NDZ.price_per_sqft.std()
NDZ
```

```python
NDZ[NDZ.z_score > 4]
```

```python
NDZ[NDZ.z_score < -4]
```

```python
outliers = NDZ[(NDZ.z_score > 4)|(NDZ.z_score < -4)]
outliers
```

```python
FNDZ = NDZ[(NDZ.z_score < 4)&(NDZ.z_score > -4)]
FNDZ

```
